```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(-0.499*math.pi, 0.499* math.pi, 0.001)

print('=================绘图==================')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x, x-x)
plt.plot(x-x, x)
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x)/x)
plt.plot(x, x/np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(np.sin(x), x)
plt.plot(np.cos(x), x)
plt.plot(np.log2(x))

```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('line')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, np.sin(x))
plt.show()
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('g')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, x, 'r')
ax.plot(x, np.sin(x), 'y')
ax.plot(x, np.cos(x), 'b')
# 绘制图例
ax.legend(('x','sin', 'cos'))
# 绘制网格
ax.grid(color='y', ls='-', lw=0.1)
# 设置刻度
ax.set_xticks([-0.5*np.pi, 0, 0.5*np.pi])
ax.set_xticklabels(['-pi/2', '0', 'pi/2'])
ax.set_yticks([-1, 0, 1])
```
